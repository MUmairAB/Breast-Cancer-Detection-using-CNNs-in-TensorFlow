# Load the model from HuggingFace for Inference

```python
#Load the model using "from_pretrained_keras" function
loaded_model = from_pretrained_keras('MUmairAB/Breast_Cancer_Detector')
```

```python
#Model summary()
loaded_model.summary()
```

```python
#Load the sample images
img0 = load_img("/content/Breast-Cancer-Detection-using-CNNs-in-TensorFlow/Images/sample3_class0.png")
img1 = load_img("/content/Breast-Cancer-Detection-using-CNNs-in-TensorFlow/Images/sample2_class1.png")
```

```python
#The loaded images are "PIL ImageFile" objects as shown below
print(type(img0)), print(type(img1))
```

```python
#Convert the PIL ImageFile to image tensor
img0 = tf.convert_to_tensor(img0)
img1 = tf.convert_to_tensor(img1)
print(img0.shape), print(img1.shape)
```

```python
#Convert the single images to batch images
img0 = tf.expand_dims(img0,axis=0)
img1 = tf.expand_dims(img1,axis=0)
print(img0.shape), print(img1.shape)
```

```python
#Make predictions
pred0 = loaded_model.predict(img0)
pred1 = loaded_model.predict(img1)
```

```python
#Let's check the resultant value and its type
type(pred0), pred0
```

```python
#Convert the "numpy.ndarray" object to a simple numebr
pred0_np = round(float(pred0))
pred1_np = round(float(pred1))
print(f"Class 0 image, Img0 is predcited as {pred0_np} Class")
print(f"Class 1 image, Img1 is predcited as {pred1_np} Class")
```
